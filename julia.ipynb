This is a simple Julia notebook. Below we'll show off the Julia `PyPlot` module.

```julia
using PyPlot
```

```julia
x = 100 * rand(500)
y = 100 * rand(500)
areas = 800 * rand(500);
```

```julia
#  Scatterplot
fig, ax = PyPlot.subplots(figsize=(10, 10))
ax.scatter(x, y, s=areas, alpha=0.5, c=areas)

ax.set_title("Scatter Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
```
